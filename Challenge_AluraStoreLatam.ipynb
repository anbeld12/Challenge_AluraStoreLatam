# **Challenge Alura Store Latam**

#### **Importación de datos**

```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head()
```

## **1. Análisis de facturación**

### **Ingreso total por tienda**

```python
# Se obtienes los ingresos totales
tienda1_ingreso_total = tienda1['Precio'].sum()
tienda2_ingreso_total = tienda2['Precio'].sum()
tienda3_ingreso_total = tienda3['Precio'].sum()
tienda4_ingreso_total = tienda4['Precio'].sum()
```

```python
# Se imprimen los ingresos totales
print("El ingreso total para la tienda 1 fue: " + tienda1_ingreso_total.astype(str))
print("El ingreso total para la tienda 2 fue: " + tienda2_ingreso_total.astype(str))
print("El ingreso total para la tienda 3 fue: " + tienda3_ingreso_total.astype(str))
print("El ingreso total para la tienda 4 fue: " + tienda4_ingreso_total.astype(str))
```

## **2. Ventas por categoría**

```python
# Se obtiene el ingreso total por categorias para cada tienda
tienda1_total_por_categoria = tienda1.groupby("Categoría del Producto")["Precio"].sum(numeric_only=True).sort_values(ascending=False)
tienda2_total_por_categoria = tienda2.groupby("Categoría del Producto")["Precio"].sum(numeric_only=True).sort_values(ascending=False)
tienda3_total_por_categoria = tienda3.groupby("Categoría del Producto")["Precio"].sum(numeric_only=True).sort_values(ascending=False)
tienda4_total_por_categoria = tienda4.groupby("Categoría del Producto")["Precio"].sum(numeric_only=True).sort_values(ascending=False)
```

## **3. Calificación promedio de la tienda**

### **Calificación general por tienda**

```python
# Se obtienen las calificaciones promedio por tienda
tienda1_calificacion = tienda1['Calificación'].mean(numeric_only=True).round(4)
tienda2_calificacion = tienda2['Calificación'].mean(numeric_only=True).round(4)
tienda3_calificacion = tienda3['Calificación'].mean(numeric_only=True).round(4)
tienda4_calificacion = tienda4['Calificación'].mean(numeric_only=True).round(4)
```

```python
# Se imprimen las calificaciones promedio por tienda
print("La calificación promedio para la tienda 1 fue: " + tienda1_calificacion.astype(str))
print("La calificación promedio para la tienda 2 fue: " + tienda2_calificacion.astype(str))
print("La calificación promedio para la tienda 3 fue: " + tienda3_calificacion.astype(str))
print("La calificación promedio para la tienda 3 fue: " + tienda4_calificacion.astype(str))
```

### **Calificación por tienda y categoria**

```python
# Se obtiene la calificacion promedio por categoria
tienda1_calificacion_por_categoria = tienda1.groupby("Categoría del Producto")["Calificación"].mean(numeric_only=True).sort_values(ascending=False).round(4)
tienda2_calificacion_por_categoria = tienda2.groupby("Categoría del Producto")["Calificación"].mean(numeric_only=True).sort_values(ascending=False).round(4)
tienda3_calificacion_por_categoria = tienda3.groupby("Categoría del Producto")["Calificación"].mean(numeric_only=True).sort_values(ascending=False).round(4)
tienda4_calificacion_por_categoria = tienda4.groupby("Categoría del Producto")["Calificación"].mean(numeric_only=True).sort_values(ascending=False).round(4)
```

## **5. Envío promedio por tienda**

```python
# Se obtiene el promedio de costo de envio por tienda
tienda1_promedio_costo_envio = tienda1['Costo de envío'].mean(numeric_only=True).round(2)
tienda2_promedio_costo_envio = tienda2['Costo de envío'].mean(numeric_only=True).round(2)
tienda3_promedio_costo_envio = tienda3['Costo de envío'].mean(numeric_only=True).round(2)
tienda4_promedio_costo_envio = tienda4['Costo de envío'].mean(numeric_only=True).round(2)

print("El costo promedio de envio para la tienda 1 es: " + tienda1_promedio_costo_envio.astype(str))
print("El costo promedio de envio para la tienda 2 es: " + tienda2_promedio_costo_envio.astype(str))
print("El costo promedio de envio para la tienda 3 es: " + tienda3_promedio_costo_envio.astype(str))
print("El costo promedio de envio para la tienda 4 es: " + tienda4_promedio_costo_envio.astype(str))
```

## **6. Visualización**

#### Porcentaje de ventas por tienda

```python
ingreso_total = {"Tienda": ["Tienda 1","Tienda 2","Tienda 3","Tienda 4"],
                 "Ingreso_total": [tienda1_ingreso_total,tienda2_ingreso_total,
                                   tienda3_ingreso_total,tienda4_ingreso_total]
                 }
df_ingreso_total_por_tienda = pd.DataFrame(ingreso_total)

df_ingreso_total_por_tienda.plot(
    kind="pie",
    y="Ingreso_total",  # Especifica la columna de valores
    labels=df_ingreso_total_por_tienda["Tienda"],  # Etiquetas para cada porción
    autopct='%1.1f%%',  # Muestra el porcentaje en cada porción
    figsize=(6, 6),     # Tamaño de la figura
    title="Porcentaje de ingreso por venta vs tienda"
)
```

#### Satisfacción clientes por tienda

```python
calificacion_promedio = {"Tienda": ["Tienda 1","Tienda 2","Tienda 3","Tienda 4"],
                         "Calificacion_promedio": [tienda1_calificacion,tienda2_calificacion,
                                                   tienda3_calificacion,tienda4_calificacion]
                         }

df_calificacion_promedio = pd.DataFrame(calificacion_promedio)

ax = df_calificacion_promedio.plot(
    kind="bar",
    x="Tienda",
    y="Calificacion_promedio",
    ylabel="Calificacion_promedio",
    xlabel="Tienda",
    figsize=(8,5),
    color="#FF6B6B",
    title="Satisfaccion de los clientes por tienda"
)

for bar in ax.patches:
    altura = bar.get_height()
    ax.annotate(
        f"{altura:.2f}",               # formato del número
        (bar.get_x() + bar.get_width() / 2, altura),
        ha="center",
        va="bottom",
        fontsize=10,
        xytext=(0, 3),                  # pequeño desplazamiento vertical
        textcoords="offset points"
    )

ax
```

```python
promedio_costo_envio = {"Tienda": ["Tienda 1","Tienda 2","Tienda 3","Tienda 4"],
                         "Costo_envio": [tienda1_promedio_costo_envio,tienda2_promedio_costo_envio,
                                                   tienda3_promedio_costo_envio,tienda4_promedio_costo_envio]
                         }

df_promedio_costo_envio = pd.DataFrame(promedio_costo_envio)

ax = df_promedio_costo_envio.plot(
    kind="bar",
    x="Tienda",
    y="Costo_envio",
    ylabel="Costo_envio",
    xlabel="Tienda",
    figsize=(8,5),
    color="#FF6B6B",
    title="Promedio costo envio por tienda"
)

for bar in ax.patches:
    altura = bar.get_height()
    ax.annotate(
        f"{altura:.2f}",               # formato del número
        (bar.get_x() + bar.get_width() / 2, altura),
        ha="center",
        va="bottom",
        fontsize=10,
        xytext=(0, 3),                  # pequeño desplazamiento vertical
        textcoords="offset points"
    )

ax
```

## **Conclusión**

Con base en los datos obtenidos y en las gráfica se ofrece una posible respuesta teniendo en cuenta las necesidades del Sr. Juan.

Si tomamos como base el ingreso total por tienda, la tienda menos eficiente sería la **tienda 4**. A su vez sería la tienda seleccionada para vender.

El objetivo de vender una tienda, es poder obtener un capital para generar otro negocio. Sin embargo, eso no implica que se tengan que descuidar las tiendas restantes, si el emprendimiento no genera los resultados esperados, tener la mayor cantidad de capital posible es el mejor salvavidas.

Analizando un poco, se podría decir que aunque no es la tienda que más vende no es la que peor está calificada por los usuarios, en respuesta a esto tenemos que tener en cuenta que el costo de envío es cubierto por el cliente. Revisando la gráfica de costo promedio de envío podemos notar que el costo promedio más bajo lo tiene la **tienda 4**, razón que podría justificar la calificación. Si un cliente paga menos por envío se va a encontrar más felíz, lo cual explicaría la calificación.

Según los datos, las tiendas 1 al 3 mantienen un orden bastante estable con respecto a sus productos más vendidos y menos vendidos. El hecho de que mantengan esta constante podría ayudar a preparse para situaciones abversas, la **tienda 4** no sigue este patrón volviendola impredecible en momentos de caos.

En conclusión, si se desea mantener una satisfacción constante y una previsión con respecto a la venta de productos, la tienda seleccionada para la venta sería la **tienda 4**.
